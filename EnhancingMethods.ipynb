```
#Ley de potencia

imc = cv2.imread("contrast2.jpeg", 0)
```

```
gam = 5.0
```

```
out = 1/(255**(gam - 1)) * imc**gam
plt.subplot(1,2,1), plt.imshow(imc, cmap = "gray") 
plt.subplot(1,2,2), plt.imshow(out, cmap = "gray") 
plt.show()
```

```
import numpy as np

def normalized_histogram(image):
  H = np.zeros((256))
  for i in range(0, image.shape[0]):
    for j in range(0, image.shape[1]):
      H[image[i,j]]+=1
  H = H/(img.shape[0]*img.shape[1])
  
  return H
```

```
h = normalized_histogram(imc)
plt.plot(h)
plt.show()
```

```
def equalize(image):
  h = normalized_histogram(image)
  A = np.zeros((256))
  A[0] = h[0]
  
  for i in range(1,256):
    A[i] = A[i - 1] + h[i]
    
  #Recorrer imagen y ecualizar
  out = np.zeros(image.shape)
  for i in range(0, image.shape[0]):
    for j in range(0, image.shape[1]):
      out[i,j] = A[image[i,j]]
  
  return out*255 # pq esta normalizada
```

```
out = equalize(imc)
plt.subplot(1,2,1), plt.imshow(imc, cmap = "gray")
plt.subplot(1,2,2), plt.imshow(out, cmap = "gray")
plt.show()
```
