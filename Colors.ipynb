```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
img = np.zeros((256,256,3), dtype = np.uint8)
```

```
img[0:32,:,:] = [255,0,0]

img[32:64,:] = [0,255,0]
img[64:96,:,:] = [0, 0, 255]
img[96:128,:,:] = [255, 0, 255]
img[128:160,:,:] = [0, 255, 255]
img[160:192,:,:] = [255, 255, 0]

img[192:224,:,:] = [255, 255, 255]

img[224:,:,:] = [255, 127.5, 127.5]
```

```
plt.imshow(img), plt.axis("off")
plt.show()
```

```
for i in range(256):
  img[i,:,:] = [255, i , i]
```

```
plt.imshow(img), plt.axis("off")
plt.show()
```

```
for i in range(256):
  img[255 - i,:,:] = [255, 255 - i , i]
```

```
plt.imshow(img), plt.axis("off")
plt.show()
```

```
import cv2 

img = np.zeros((360,256,3), np.uint8)

for i in range(180):
    img[ i*2:2*i + 2, :, :] = [i, 127, 127] # color , saturacion, oscuridad
    
cvt = cv2.cvtColor(img, cv2.COLOR_HSV2RGB)
```

```
plt.subplot(1,2,1), plt.imshow(img), plt.axis("off")
plt.subplot(1,2,2), plt.imshow(cvt), plt.axis("off")
plt.show()
```
